```python

import cv2
import numpy as np
import matplotlib.pyplot as plt


IMAGE_PATH = r"C:\Users\zaid\Downloads\archive (4)\twenty jordanian dinar\e94afae6-e70d-4820-8bb3-84e4aaf58f88.jpg"


img = cv2.imread(IMAGE_PATH)
if img is None:
    raise FileNotFoundError("Image not found. Check the path.")


rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def show(title, image, cmap=None):
    plt.figure(figsize=(5,5))
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

show("Original Image (RGB)", rgb)


print("Image shape (Height, Width, Channels):", rgb.shape)
print("Pixel value at (50,50):", rgb[50,50])


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show("Grayscale Image", gray, cmap="gray")


hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

show("HSV Image", hsv)


h, s, v = cv2.split(hsv)
show("Hue Channel", h, cmap="gray")
show("Saturation Channel", s, cmap="gray")
show("Value Channel", v, cmap="gray")


plt.figure(figsize=(6,4))
plt.hist(gray.ravel(), bins=256, range=[0,256])
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()


equalized = cv2.equalizeHist(gray)

show("Histogram Equalization", equalized, cmap="gray")


plt.figure(figsize=(6,4))
plt.hist(equalized.ravel(), bins=256, range=[0,256])
plt.title("Equalized Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()

avg_blur = cv2.blur(gray, (5,5))
show("Average Blur", avg_blur, cmap="gray")


gaussian_blur = cv2.GaussianBlur(gray, (5,5), 0)
show("Gaussian Blur", gaussian_blur, cmap="gray")


median_blur = cv2.medianBlur(gray, 5)
show("Median Blur", median_blur, cmap="gray")


edges = cv2.Canny(gray, 100, 200)
show("Edge Detection (Canny)", edges, cmap="gray")


print("Image Processing Fundamentals Completed Successfully!")
```
